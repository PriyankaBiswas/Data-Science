## Challenge: Evaluate an experiment analysis 

*Now it's time to flex your critical evaluation skills. Read the following descriptions of an experiment and its analysis, identify the flaws in each, and describe what you would do to correct them.*

** **

**The Sith Lords are concerned that their recruiting slogan, "Give In to Your Anger," isn't very effective. Darth Vader develops an alternative slogan, "Together We Can Rule the Galaxy." They compare the slogans on two groups of 50 captured droids each. In one group, Emperor Palpatine delivers the "Anger" slogan. In the other, Darth Vader presents the "Together" slogan. 20 droids convert to the Dark Side after hearing Palpatine's slogan, while only 5 droids convert after hearing Vader's. The Sith's data scientist concludes that "Anger" is a more effective slogan and should continue to be used.** 

*Although I don't have much experience of the Star Wars history, I did a quick search. I am assuming the following: 

*- Droids are robots, and I assume they don't have emotions. *

*- Emperor Palpatine is a orator and a senator, apparently "charismatic."*

*- Darth Vader is an iconic villain.*

**Flaws**

- This experimentation is biased, most probably has a contextual or an observer bias. 

    - Contextual bias, because Palpatine is a better and more charismatic leader. The droids are more likely to convert and that's what happens. 
    
    - Also, contextual bias because these droids are captured, and the slogan "Give in to your anger" may have a greater effect as it appeals to their current state and associated states, than that of "Ruling the Galazy". 
    
**Alternatives** 
 
- Both have the same moderator, either Palpatine or Darth Vader. 

- Have similar conditions semantically - "Give in to your anger" has a completely different message than "Conquering the Galaxy"  


** **

**In the past, the Jedi have had difficulty with public relations. They send two envoys, Jar Jar Binks and Mace Windu, to four friendly and four unfriendly planets respectively, with the goal of promoting favorable feelings toward the Jedi. Upon their return, the envoys learn that Jar Jar was much more effective than Windu: Over 75% of the people surveyed said their attitudes had become more favorable after speaking with Jar Jar, while only 65% said their attitudes had become more favorable after speaking with Windu. This makes Windu angry, because he is sure that he had a better success rate than Jar Jar on every planet. The Jedi choose Jar Jar to be their representative in the future.**


**Flaws**

- This experimentation has the Simpson's paradox. To begin with, Jar Jar went to 4 friendly planets, while Mace Windu went to 4 unfriendly planets. Unfrendly planets will most likely be difficult to achieve favorable conversion. However, as stated, Windu had a better success rate on every planet, therefore this is a classic case Simpson's Paradox. The friendliness is the lurking variable here.
    
**Alternatives** 
 
- To begin with, the sampling distribution is not well controlled. Ideally, each envoy should have 2 friendly and 2 unfriendly planets. 

- If current experimental conditions are kept, success rates for every planet should be calculated rather than an average of success rates. Secondly, success rates in the unfriendly planets should be weighted higher than the friendly planets.

**A company with work sites in five different countries has sent you data on employee satisfaction rates for workers in Human Resources and workers in Information Technology. Most HR workers are concentrated in three of the countries, while IT workers are equally distributed across worksites. The company requests a report on satisfaction for each job type. You calculate average job satisfaction for HR and for IT and present the report.**

**Flaws**
- Sampling bias in the HR: if the HR job satisfaction rates are concentrated on three countries, an average does not sbow an overall job satisfaction in the category of HR. 

- IT is equally distributed, hence an average over all IT jobs might be appropriate. 

    - Even though it is equally distributed, there might be contextual bias such as work conditions, work hours, or benefits in each country. 
    

**Alternatives**

- do a country-wise satisfaction calculation.

**When people install the Happy Days Fitness Tracker app, they are asked to "opt in" to a data collection scheme where their level of physical activity data is automatically sent to the company for product research purposes. During your interview with the company, they tell you that the app is very effective because after installing the app, the data show that people's activity levels rise steadily.**


**Flaws**

- Sampling bias. The finding is only from those who opted in, hence it is not appropriate to generalize the use of all users. 

- One good point is that it does not show abrupt rise in activity levels, as it is expected that users will overuse the app after installing. A steady increase might be a sign that the app is performing well in keeping the users active and motivated.  

**Alternative**
- rephrase the finding to justify that the finding is only for those who opted in.

**To prevent cheating, a teacher writes three versions of a test. She stacks the three versions together, first all copies of Version A, then all copies of Version B, then all copies of Version C. As students arrive for the exam, each student takes a test. When grading the test, the teacher finds that students who took Version B scored higher than students who took either Version A or Version C. She concludes from this that Version B is easier, and discards it.**

**Flaws** 

*Assuming that the three versions were similar, had same numbers of copies, and all students started and ended the exams at the same time* 

- Sampling bias, because the A and C versions were distributed to students those who came early or came late. It's possible that these students were more anxious than those who received the version B. Those who came early and got exam A could have exam fright and are more anxoious in general. Those who came late got Exam C and were anxious for the time limitation. 

    - This also supports my assumption about the three versions being similar. If the B students do better, then they were more relaxed and performed well. 
    
    
**Alternatives**

- Interleaf each version (A,B,C) and distribute to students. 
- Start the exam only after everyone got a copy. 
- Do not discard!
