## Drill1: A/B Tests: Getting Testy

For each of the following questions, outline how you could use an A/B test to find an answer. Be sure to identify all five key components of an A/B test we outlined above.

- Does a new supplement help people sleep better?
- Will new uniforms help a gym's business?
- Will a new homepage improve my online exotic pet rental business?
- If I put 'please read' in the email subject will more people read my emails?


Five key components of A/B Testing: 
1. Versions
2. Sample
3. Hypothesis
4. Outcome
5. Other measured variables



**--> Does a new supplement help people sleep better?**

1. **Versions**
    
    *Test Version:* receives the supplement, monitor sleep pattern, ask qualitative questions
    
    *Control Version:* receives the Pacebo, and experiemental conditions are the same as the test version. 
   
    
2. **Sample** 

Random sampling into the two groups as mentioned above. Should have similar age, gender, pre-treatment health conditions.


3. **Hypothesis**

*Null hypothesis:* there is no difference in the sleep patterns after receiving the supplements or the placebo.  

*Alternate hypothesis:* People receiving the supplemen will sleep better than the onese receiving the placebo.  

4. **Outcome** 

Key metrics: 
    1. How long each participant is sleeping 
    2. If they have sound sleep or they have disturbances in sleep
    3. If they are feeling drowsy or energetic during the day.
    4. Self-report: if they think it's working 


5. **Other measured variables**

Demographic information of the population: Age, gender, work habit, if they exercise during the day etc. 





**--> Will new uniforms help a gym's business?**

*I am assuming uniforms on the gym workers/instructors and not on the customers :-)

1. **Versions**
    
    *Test Version:* wear new uniforms during work hours and in customer-facing interactions for a predefined trial period
    
    *Control Version:* doesn't wear uniforms or have the old uniforms. 
   
    
2. **Sample** 

Existing members and new members. 


3. **Hypothesis**

*Null hypothesis:* there is no difference in the business and customer satisfaction.  

*Alternate hypothesis:* New uniforms increases customer satisfaction.  

4. **Outcome** 

Key metrics: 
    1. Number of new customer gain during the trial period. 
    2. Customers' duration of each session
    3. Customers' number of visit 
    4. end-of-trial customer survey


5. **Other measured variables**

Demographic information of the population: Age, gender, gym habit, if they exercise during the day or night, and their qualitative judgment about the uniforms etc.
